```python
import numpy as np
import matplotlib.pyplot as plt

# LU decomposition comparing theoretical and experimental
sizes=[200,300,500,1000,2000, 3000, 5000]
times=[0.0420000000, 0.0980000000, 0.2890000000, 2.2430000000, 20.8560000000, 60.3170000000, 319.2160000000] # time we got for each size using our algorithm

lu = [n**3 * 10**-6 for n in sizes] # theoretical LU time: O(n^3) --> size^3 * 10^-6. 10^-6: constant factor for each operation

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, label="Experimental Time Complexity for LU Decomposition", marker='o')
plt.plot(sizes, lu, label="Theoretical Time Complexity (O(n^3)) for LU Decomposition", linestyle='--')

plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel("Matrix Size (n)")
plt.ylabel("Time / Complexity")
plt.title("Matrix Size vs Complexity for LU Decomposition")
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

```python
#  Naive matrix multiplication comparing theoretical and experimental
sizes=[200,300,500,1000,2000, 3000]
times=[0.049000, 0.171000, 0.615000, 12.895000, 162.258000 , 480.831000] # time we got for each size using our algorithm

naive = [n**3 * 10**-6 for n in sizes] # theoretical Naive mult time: O(n^3) --> size^3 * 10^-6. 10^-6: constant factor for each operation

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, label="Experimental Time Complexity for Naive Matrix Multiplication", marker='o')
plt.plot(sizes, naive, label="Theoretical Time Complexity (O(n^3)) for Naive Matrix Multiplication", linestyle='--')

plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel("Matrix Size (n)")
plt.ylabel("Time / Complexity")
plt.title("Matrix Size vs Complexity for Naive Multiplication")
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Strassen matrix multiplication comparing theoretical and experimental
sizes=[10, 20, 50, 80, 100, 200, 300, 500, 520]
times=[0.009000, 0.116000, 0.312000 , 2.009000, 1.979000, 13.188000, 98.234000, 73.913000, 725.707000] # time we got for each size using our algorithm

strassen = [n**np.log2(7) * 10**-6 for n in sizes] # theoretical Strassen mult time: O(n^(log2(7))) --> size^(log2(7)) * 10^-6. 10^-6: constant factor for each operation

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, label="Experimental Time Complexity for Strassen Matrix Multiplication", marker='o')
plt.plot(sizes, strassen, label="Theoretical Time Complexity (O(n^2.81)) for Strassen Matrix Multiplication", linestyle='--')

plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel("Matrix Size (n)")
plt.ylabel("Time / Complexity")
plt.title("Matrix Size vs Complexity for Strassen Multiplication")
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Matrix inversion using LU comparing theoretical and experimental
sizes=[200, 300, 500, 1000, 2000, 3000]
times=[0.127000, 0.261000, 1.118000, 8.258000, 70.940000, 225.994000] # time we got for each size using our algorithm
LU_complexity = [n**3 * 10**-6 for n in sizes]  # theoretical LU inversion time: O(n^3) --> size^3 * 10^-6. 10^-6: constant factor for each operation 

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, label="Experimental Time Complexity for LU Matrix Inversion", marker='o')
plt.plot(sizes, LU_complexity, label="Theoretical Time Complexity O(n^3) for LU Matrix Inversion", linestyle='--')

plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel("Matrix Size (n)")
plt.ylabel("Time / Complexity")
plt.title("Matrix Size vs Complexity for LU Inversion")
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Matrix inversion using Strassen/Strassen comparing theoretical and experimental
sizes=[20, 30, 40 , 50, 80, 100, 200, 300, 500, 520]
times=[0.135000, 0.071000, 0.314000, 0.352000, 1.790000, 1.951000, 10.896000, 77.178000, 79.945000, 645.742000 ] # time we got for each size using our algorithm

strassen_complexity = [(10**-6 * n**np.log2(7))*(np.log2(n)) for n in sizes]  # theoretical Strassen/Strassen inversion time: O(n^log2(7)) --> size^log2(7) * 10^-6. 10^-6: constant factor for each operation

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, label="Experimental Time Complexity for Strassen/Strassen Matrix Inversion", marker='o')
plt.plot(sizes, strassen_complexity, label="Theoretical Time Complexity O(n^log2(7).log2(7)) for Strassen/Strassen Matrix Inversion", linestyle='--')

plt.xscale('log', base=2)
plt.yscale('log')
plt.xlabel("Matrix Size (n)")
plt.ylabel("Time / Complexity")
plt.title("Matrix Size vs Complexity for Strassen/Strassen Inversion")
plt.legend()
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```
